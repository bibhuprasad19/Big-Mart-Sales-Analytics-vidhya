```
total_data=train_data[['Item_Weight','Item_Visibility','Item_MRP']]
total_data
```

```
final_data=pd.DataFrame({'Item_Fat_Content':train_data['Item_Fat_Content'],'Item_Type':train_data['Item_Type'],'Outlet_Identifier':train_data['Outlet_Identifier'],'Outlet_Establishment_Year':train_data['Outlet_Establishment_Year']
                                ,	'Outlet_Location_Type':train_data['Outlet_Location_Type'],	'Outlet_Type':train_data['Outlet_Type']})
final_data
                         
                         
                
```

```
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import numpy as np
Labelencoder=LabelEncoder()

final_data['Item_Fat_Content']=Labelencoder.fit_transform(final_data['Item_Fat_Content'])
final_data['Item_Type']=Labelencoder.fit_transform(final_data['Item_Type'])
final_data['Outlet_Establishment_Year']=Labelencoder.fit_transform(final_data['Outlet_Establishment_Year'])
final_data['Outlet_Identifier']=Labelencoder.fit_transform(final_data['Outlet_Identifier'])
#final_data['Outlet_Size']=Labelencoder.fit_transform(final_data['Outlet_Size'])
final_data['Outlet_Location_Type']=Labelencoder.fit_transform(final_data['Outlet_Location_Type'])
final_data['Outlet_Type']=Labelencoder.fit_transform(final_data['Outlet_Type'])
```

```
final_data
```

```
one_hot=OneHotEncoder(handle_unknown='ignore')
one_hot_df=pd.DataFrame(one_hot.fit_transform(final_data).toarray())

one_hot_df
```

```
total_data_new=total_data.join(one_hot_df,lsuffix='_left',rsuffix='_right')
total_data_new.head()
```

```
total_data_new.isnull().sum()
```

```
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='mean')
total_data_abs=imp.fit_transform(total_data_new)
total_data_abs
```

```
from sklearn.model_selection import train_test_split
X=total_data_abs
Y=train_data['Item_Outlet_Sales']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()


X_train_scaled=sc_x.fit_transform(X_train)
X_test_scaled=sc_x.fit_transform(X_test)
```

```
from sklearn.linear_model import LinearRegression,ElasticNetCV,LassoCV,RidgeCV
import xgboost 
from sklearn.ensemble import AdaBoostRegressor,RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import GridSearchCV
```

```
#linear_regression
lin_regressor=LinearRegression()
lin_regressor.fit(X_train_scaled,Y_train)

#elasticnet
elast_regressor=ElasticNetCV()#alpha=1.0,l1_ratio=0.5
elast_regressor.fit(X_train_scaled,Y_train)

#lasso
lasso_regressor=LassoCV()
lasso_regressor.fit(X_train_scaled,Y_train)

#Ridge
ridge_regressor=RidgeCV()
ridge_regressor.fit(X_train_scaled,Y_train)

#xgboost
xgb_regressor=xgboost.XGBRegressor(max_depth=5,learning_rate=0.08,n_estimators=150)
xgb_regressor.fit(X_train_scaled,Y_train)

#adaboost
ada_regressor=AdaBoostRegressor(n_estimators=200,learning_rate=3,loss='linear')
ada_regressor.fit(X_train_scaled,Y_train)

#decisionTree
tree_regressor=DecisionTreeRegressor(max_depth=250)
tree_regressor.fit(X_train_scaled,Y_train)

#randomforest
forest_regressor=RandomForestRegressor(n_estimators=100,max_depth=10,random_state=0)
forest_regressor.fit(X_train_scaled,Y_train)

#KNN
knn_regressor=KNeighborsRegressor(n_neighbors=7,weights='uniform')
knn_regressor.fit(X_train_scaled,Y_train)

#SVR
svr_regressor=SVR(gamma=0.05,kernel='linear',C=5)#
svr_regressor.fit(X_train_scaled,Y_train)





```

```
y_pred_lin = lin_regressor.predict(X_test_scaled)
y_pred_elast=elast_regressor.predict(X_test_scaled)
y_pred_lasso=lasso_regressor.predict(X_test_scaled)
y_pred_ridge=ridge_regressor.predict(X_test_scaled)
y_pred_xgb = xgb_regressor.predict(X_test_scaled)
y_pred_ada = ada_regressor.predict(X_test_scaled)
y_pred_tree = tree_regressor.predict(X_test_scaled)
y_pred_forest = forest_regressor.predict(X_test_scaled)
y_pred_knn = knn_regressor.predict(X_test_scaled)
y_pred_svr = svr_regressor.predict(X_test_scaled)
```

```
from sklearn import metrics 

print('Root Mean Squared Error of linear:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_lin)))
print('Root Mean Squared Error of Elasticnet:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_elast)))
print('Root Mean Squared Error of Lasso:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_lasso)))
print('Root Mean Squared Error of ridge:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_ridge)))

print('Root Mean Squared Error of xgb:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_xgb)))
print('Root Mean Squared Error of ada:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_ada)))
print('Root Mean Squared Error of tree:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_tree)))
print('Root Mean Squared Error of forest:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_forest)))
print('Root Mean Squared Error of knn:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_knn)))
print('Root Mean Squared Error of svr:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred_svr)))
```
